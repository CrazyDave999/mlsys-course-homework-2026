## Part 1: Understanding the 2D Cumulative Sum Operation

### What is Cumulative Sum (Prefix Sum)?

Cumulative Sum, also known as Prefix Sum or Scan, is a fundamental operation in parallel computing. It transforms a sequence of numbers into another sequence where each element is the sum of all preceding elements in the input.

For an input [a, b, c, d], the output is [a, a+b, a+b+c, a+b+c+d]. This operation is a key building block for many complex parallel algorithms.

### The Core Task: Row-wise 2D Cumulative Sum

Your objective is to implement a parallel algorithm that computes the inclusive scan independently for each row of a given 2D matrix. This structure means that while each row's calculation is independent of other rows, the computation within each row requires a sophisticated parallel strategy.

#### Mathematically:

The operation can be defined mathematically as follows. For each row i and element j in the output matrix Out:

```
  Out[i][j] = Σ (from k=0 to j) A[i][k]
```

### Why a Parallel Scan on the GPU?
A simple for loop is inherently sequential, as each step i depends on the result from step i-1. This approach cannot leverage the massive parallelism of a GPU.

A **parallel scan algorithm** breaks this dependency by performing the summation in a tree-like manner across multiple steps. This allows thousands of GPU threads to work together on the problem, leading to massive speedups over a sequential CPU implementation.

For a detailed explanation of how to implement this in parallel, please review the following resources:

* NVIDIA Official Blog Post: Parallel Prefix Sum (Scan) with CUDA [site](https://developer.nvidia.com/gpugems/gpugems3/part-vi-gpu-computing/chapter-39-parallel-prefix-sum-scan-cuda)
* Wikipedia: Prefix Sum [site](https://en.wikipedia.org/wiki/Prefix_sum)
